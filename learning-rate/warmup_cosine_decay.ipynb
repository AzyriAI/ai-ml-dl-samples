```python
import yaml
import numpy as np
import matplotlib.pyplot as plt
from warmup_cosine_decay import lr_warmup_cosine_decay
from munch import munchify

cfg = 'config.yaml'
with open(cfg) as f:
  parameters = munchify(yaml.load(f, Loader=yaml.SafeLoader))

lrs = []
epochs = range(0, parameters.epochs)
for epoch in epochs:
  lr, de = lr_warmup_cosine_decay(epoch=epoch, parameters=parameters)
  lrs.append(lr)
plt.plot(epochs, lrs)
plt.show()
```
